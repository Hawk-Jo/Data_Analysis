```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
# 데이터 불러오기
folder_path = "./data"
# 폴더 내 모든 CSV 파일 가져오기
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# CSV 파일 병합
df_list = [pd.read_csv(file) for file in csv_files]  # 각 CSV를 데이터프레임으로 변환
merged_df = pd.concat(df_list, ignore_index=True)  # 병합 후 인덱스 재설정
```
